```python
team_distribution = data['Team'].value_counts()

print("Employee count per team:")
print(team_distribution)
```

```python
team_percentage = (team_distribution / len(data)) * 100

print("\nPercentage split of employees per team:")
print(team_percentage)
```

```python
team_analysis = pd.DataFrame({
    'Team': team_distribution.index,
    'Employee Count': team_distribution.values,
    'Percentage': team_percentage.values
})

team_analysis = team_analysis.sort_values(by='Employee Count', ascending=False).reset_index(drop=True)

print("\nFinal team analysis (sorted by employee count):")
print(team_analysis)
```

```python
import pandas as pd

data = pd.read_csv(r"C:\Users\PRAVEEN\Downloads\myexcel - myexcel.csv.csv")


print(data.head())
```

```python
bins = [0, 20, 30, 40, 50, 60, 100]
labels = ['Under 20', '20-30', '31-40', '41-50', '51-60', 'Above 60']

data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

age_group_distribution = data['Age Group'].value_counts()

print("Employee count per age group:")
print(age_group_distribution)

predominant_age_group = age_group_distribution.idxmax()
print(f"\nThe predominant age group is: {predominant_age_group}")
```

```python
team_salary_expenditure = data.groupby('Team')['Salary'].sum()

highest_salary_team = team_salary_expenditure.idxmax()
highest_salary_team_expenditure = team_salary_expenditure.max()

print("Team salary expenditures:")
print(team_salary_expenditure)
print(f"\nThe team with the highest salary expenditure is: {highest_salary_team}")
print(f"Total salary expenditure for {highest_salary_team}: {highest_salary_team_expenditure}")
```

```python
position_salary_expenditure = data.groupby('Position')['Salary'].sum()

highest_salary_position = position_salary_expenditure.idxmax()
highest_salary_position_expenditure = position_salary_expenditure.max()

print("\nPosition salary expenditures:")
print(position_salary_expenditure)
print(f"\nThe position with the highest salary expenditure is: {highest_salary_position}")
print(f"Total salary expenditure for {highest_salary_position}: {highest_salary_position_expenditure}")
```

```python
missing_values = data[['Age', 'Salary']].isnull().sum()

print("Missing values in Age and Salary columns:")
print(missing_values)
```

```python
correlation = data[['Age', 'Salary']].corr().iloc[0, 1]

print(f"\nThe correlation coefficient between Age and Salary is: {correlation}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Age'], y=data['Salary'], alpha=0.7, color='blue')

sns.regplot(x=data['Age'], y=data['Salary'], scatter=False, color='red', ci=None)

plt.title("Scatter Plot of Age vs. Salary", fontsize=16)
plt.xlabel("Age", fontsize=14)
plt.ylabel("Salary", fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```
